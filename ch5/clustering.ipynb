```
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

Read the data file and remove unused columns

```
df_source = pd.read_csv("data.csv")
df = df_source

columns = ["case_no", "onset_date", "confirmation_date", "gender", "status", "citizenship_zh", "district", "district_en", "classification", "lat", "long", "address"]
for col in columns:
  df = df.drop(col, 1)

df
```

Perform clustering

```
clustering = DBSCAN(eps=5, min_samples=3)
result = clustering.fit_predict(df)
max(result)
```

Save the clustering result

```
df_source['result'] = result
df_source.to_csv("result.csv", encoding="utf-8", index=False)
```

Perform Principal Component Analysis and save the result

```
pca = PCA(n_components=5)
pca_result = pd.DataFrame(pca.fit_transform(df))

pca_result['result'] = result
pca_result.to_csv("pca.csv", encoding="utf-8", index=False)
```

Take a look at pricipal components

```
df_components = pd.DataFrame(pca.components_).transpose()
df_components['columns'] = df.columns.values
df_components.to_csv("components.csv", encoding="utf-8", index=False)
df_components
```
